```python
import numpy as np
import matplotlib.pyplot as plt
```

Given that a new approach is required to avoid the problem of no-normal direction burning (but radial direction), this notebook will try to address that problem. Elliptical grid will be the used mapping to convert a squared array to a circular disposition (more info at https://arxiv.org/pdf/1509.06344.pdf).

```python
# number of points per axis (total points will be N^N)
N = 100

# square horizontal (x) and vertical (y) axis
x = np.linspace(-1,1,N)
y = -np.linspace(-1,1,N)

# square meshgrid
X, Y = np.meshgrid(x,y)
```

```python
# circle meshgrid (from squared meshgrid)
U = X*np.sqrt(1-Y**2/2)
V = Y*np.sqrt(1-X**2/2)
```

```python
# plot the circular mapping from the squared array
fig, ax = plt.subplots(1, 2, figsize=(32,16))
ax[0].plot(X,Y,'k.')
ax[1].plot(U,V,'k.')
[ax[i].axis('equal') for i in range(len(ax))];
[ax[i].axis('off') for i in range(len(ax))];
```

## Other possible approach

If a squared array is known, the elements whose coordinates are inside an imaginary unitary circle may also be selected as the possible discretization. At first, the previous one will be used. However, this is another approach that may be used.

```python
# boolean matrix with the coordinates that belong to the circle
circArray = np.sqrt(X**2+Y**2) < 1
```

```python
# plot the squared array restricted to the unity circle
fig, ax = plt.subplots(1, figsize=(8,8), dpi=200)
ax.plot(X[circArray],Y[circArray],'k.',ms=1)
ax.axis('equal')
ax.axis('off');
```
